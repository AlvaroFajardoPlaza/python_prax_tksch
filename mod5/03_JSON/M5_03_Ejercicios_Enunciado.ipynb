```python
# Vamos a crear el diccionario, empezando por las partes más pequeñas, hasta llegar a las partes más grandes.

# 1º Creamos a cada persona<dict>, con sus atributos y su experiencia<array>
experiencia_Juan: list = ['Gestión', 'Finanzas', 'Bases de datos']
experiencia_Elena: list = ['JavaScript', 'Python']
experiencia_Luis: list = ['Django', 'Flask', 'Pyramid']

juan_data: dict = { 
    'Nombre': 'Juan',
    'Edad': 28,
    'Residencia': 'Madrid',
    'Horas de Proyecto': 3500,
    'Experiencia': experiencia_Juan
    }
elena_data: dict = {
    'Nombre': 'Elena',
    'Edad': 26,
    'Residencia': 'Madrid',
    'Horas de Proyecto': 500,
    'Experiencia': experiencia_Elena
    }
luis_data: dict = {
    'Nombre': 'Luis',
    'Edad': 31,
    'Residencia': 'Barcelona',
    'Horas de Proyecto': 1100,
    'Experiencia': experiencia_Luis
}

# 2º Generamos al Jefe de Proyecto<dict> y la lista de Empleados<array>
empleados: list = [elena_data, luis_data]

# 3º Generamos el diccionario final que contiene toda la info: jefe de proyecto + array de empleados
proyecto_beta: dict = { 'jefe_proyecto': juan_data, 'empleados': empleados }
print(proyecto_beta)
```

3) A veces os encontraréis con JSON que tendréis que modificar. Para ello tenéis que decodificarlos, realizar las modificaciones pertinentes y volver a codificarlo para dejarlo como JSON de nuevo. En el siguiente ejemplo os habéis dado cuenta de que hay algunos errores:
* A Superman le falta como poder "Volar"
* En Batman, la edad es 35, no 350
* En Batman, le sobra el poder de "Rayos en los ojos"
* En Wonder Woman le falta el poder "Lazo de la verdad"
* Después de corregir todo esto, transforma estos datos en un JSON String

```python
import json

# NO TOCAR NADA DE ESTE BLOQUE
superheroes = {
	"nombreEquipo": "Super Hero Squad",
	"ciudad": "Metro City",
	"formado": 2016,
	"baseSecreta": "Super Tower",
	"activo": "Si",
	"miembros": [
		{
			"nombre": "SuperMan",
			"edad": 29,
			"identidadSecreta": "Clart Kent",
			"poderes": [
				"Super fuerza",
				"Super velocidad",
				"Rayos en los ojos"
			]
		},
		{
			"nombre": "Batman",
			"edad": 350,
			"identidadSecreta": "Bruce Wayne",
			"poderes": [
				"Detective",
				"Dinero",
				"Rayos en los ojos"
			]
		},
		{
			"nombre": "Wonder Woman",
			"edad": 900,
			"identidadSecreta": "Diana de Temiscira",
			"poderes": [
				"Super fuerza",
				"Super velocidad"
			]
		}
	]
}

# 1º Comprobamos el tipo con el que vamos a trabajar
print(type(superheroes), superheroes)

# 2º Realizamos las modificaciones solicitadas:
superheroes["miembros"][0]["poderes"].append("Volar")
superheroes["miembros"][1]["edad"] = 35
superheroes["miembros"][1]["poderes"].pop()
superheroes["miembros"][2]["poderes"].append("Lazo de la Verdad")

print(superheroes)

# 3º Ahora que lo tenemos modificado, podemos volver a convertir el archivo a JSON
json_superheroes = json.dumps(superheroes, indent=4)
print("\n ---- SUPERHEROES JSON ----")
print(type(json_superheroes))
print(json_superheroes)
```

4) En base al ejercicio anterior, modifica la estructura de super para lograr que miembros tenga dos ramas: "miembrosActivos" y "miembrosInactivos", donde cada una de estas ramas, almacenen los héroes que están en activo y los que no. En este caso, introduce a SuperMan y Wonder Woman en la lista de activos y a Batman en la de Inactivos. Esta modificación puedes hacerla como quieras, o bien programando las estructuras de datos e ir componiéndolo o bien cogiendo el JSON y modificándolo sobre él mismo. Al finalizar puedes comprobar tu JSON en un visualizador online (http://jsonviewer.stack.hu/)

```python
# NO TOCAR NADA DE ESTE BLOQUE
superheroes = {
    "nombreEquipo": "Super Hero Squad",
    "ciudad": "Metro City",
    "formado": 2016,
    "baseSecreta": "Super Tower",
    "activo": "Si",
    "miembros": [
        {
            "nombre": "SuperMan",
            "edad": 29,
            "identidadSecreta": "Clart Kent",
            "poderes": [
                "Super fuerza",
                "Super velocidad",
                "Rayos en los ojos",
                "Volar"
            ]
        },
        {
            "nombre": "Batman",
            "edad": 35,
            "identidadSecreta": "Bruce Wayne",
            "poderes": [
                "Detective",
                "Dinero"
            ]
        },
        {
            "nombre": "Wonder Woman",
            "edad": 900,
            "identidadSecreta": "Diana de Temiscira",
            "poderes": [
                "Super fuerza",
                "Super velocidad",
                "Lazo de la Verdad"
            ]
        }
    ]
}


# Sobre nuestro archivo JSON, queremos que "miembros" sea un dict con dos ramas: "miembros activos" <array> y "miembros inactivos" <array>
print(superheroes)

# Vamos a extraer los miembros con un bucle for y a introducirlos en nuesvas listas:
miembros_activos = []
miembros_inactivos = []

for hero in superheroes["miembros"]:
    if hero["nombre"] == "Batman":
        miembros_inactivos.append(hero)
    else:
        miembros_activos.append(hero)

print("Los superheroes en activo son: ",miembros_activos,"\nLos superheroes en inactivo son: ", miembros_inactivos)

# Ya tenemos nuestras dos listas. Ahora hay que incluirlas dentro de nuestro dict de superheroes:
superheroes.pop("miembros") # Eliminamos esta key
miembros = [ miembros_activos, miembros_inactivos ]
superheroes["miembros"] = {"miembros_activos": miembros_activos, "miembros_inactivos": miembros_inactivos} # regeneramos la key con nuestros listados de valores
print("\n------", superheroes, type(superheroes))

# Transformamos a JSON str para visualizar nuestro resultado
json_superheroes = json.dumps(superheroes, indent=4)
print(json_superheroes, type(json_superheroes))
```

5) En el siguiente código, accedemos a un JSON de forma remota, a partir de la respuesta, realizar lo siguiente:
* Mostrar el tipo de dato que se ha recibido
* Mostrar los datos recibidos
* Mostrar el número de personas que se encuentran actualmente en el espacio
* Realizar un bucle que recorra a todas esas personas y muestre nombre y nave en la que se encuentra.

```python
import requests

# API que nos comunica cuantas personas se encuentran actualmente en el espacio
response = requests.get("http://api.open-notify.org/astros.json")
data = response.json()

print(data, type(data))

# Función que nos devuelve el número de personas que hay en el espacio:
def num_people_in_space(dict):
    # De nuestro dict accedemos al valor de la key "number"
    print("El número de personas que hay en el espacio en este momento es: {}.".format(dict["number"]))

# Función para mostrar los datos de las personas y su nave:
def show_astronaut_data(dict):
    print("\nDatos de los astronautas:")
    for astronaut in dict["people"]:
        print("Nombre: {} --- Nave: {}".format(astronaut["name"], astronaut["craft"]))

num_people_in_space(data)
show_astronaut_data(data)


```
